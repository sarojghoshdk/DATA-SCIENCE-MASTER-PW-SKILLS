Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

ANS:- In Pandas, a DataFrame is a two-dimensional labeled data structure that can store data of different types (e.g., numerical, string, boolean) in a tabular form. It consists of rows and columns, similar to a SQL table or a spreadsheet.where each column can be considered as a Pandas Series.

On the other hand, a Series in Pandas is a one-dimensional labeled array that can contain data of any 
    data type, including integers, strings, and floating-point numbers. It is similar to a column in a 
    spreadsheet or a SQL table.
         

The main difference between a DataFrame and a Series is that a DataFrame is a two-dimensional data 
    structure with rows and columns, while a Series is a one-dimensional data structure with only one column.
    

Here's an example to illustrate the difference between a DataFrame and a Series:

```python
# example of Series
import pandas as pd
data = {'Name' : ['Alice','Bob','Claire'],
        
        'Age' : [25,30,27],
                  
        'Gender' : ['Female','Male','Female']}

df = pd.Series(data)
df                
```

```python
# example of DataFream
import pandas as pd
data = {'Name' : ['Alice','Bob','Claire'],
        
        'Age' : [25,30,27],
                  
        'Gender' : ['Female','Male','Female']}

df = pd.DataFrame(data)
df                 
```

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

ANS:- 

In Pandas, a Series object is mutable, which means its values can be modified after creation. You can assign new values to specific elements or slices of a Series using indexing or attribute assignment.

On the other hand, a DataFrame object is also mutable. You can modify the values of individual cells, add or remove columns, and perform various transformations on the DataFrame.

However, the Panel object is not mutable. Panel was a 3D data structure in earlier versions of Pandas but has been deprecated since version 0.25.0. 

--> Series: Mutable

--> DataFrame: Mutable

--> Panel: Deprecated and not mutable

In a mutable object, the contents of the object can be modified after it is created. This means 
     that you can add, remove, or update elements in a Series or DataFrame after they are created.
