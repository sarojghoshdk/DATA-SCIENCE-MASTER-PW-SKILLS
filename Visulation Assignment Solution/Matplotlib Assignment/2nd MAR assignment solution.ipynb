# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

# ANS:- 

# Matplotlib is a powerful data visualization library for Python that is used to create static, animated, and interactive visualizations in Python. It provides a wide range of plotting options and tools to create a variety of high-quality visualizations.

# Matplotlib is a cross-platform, data visualization and graphical plotting library for Python and its numerical extension NumPy.It simplifies the process of creating visualizations by providing a consistent interface for different types of plots.


# Matplotlib is used for several purposes such as:
    
1. Creating line plots, scatter plots, bar plots, and histograms to explore and visualize data

2. Creating publication-quality figures and plots

3. Generating visualizations for reports and presentations

4. Creating custom visualizations with unique specifications and designs

5. Visualizing machine learning models and results

# Here are five types of plots that can be created using the Pyplot module of Matplotlib:

1. Line plots: These plots display data as a series of points connected by a line. They are commonly used to display trends and 
   changes in data over time.

2. Scatter plots: These plots display data as a set of points, where each point represents a single observation or data point. 
   They are commonly used to identify patterns and relationships between variables.

3. Bar plots: These plots display data as a set of rectangular bars, where the height of each bar represents the value of the 
   corresponding data point. They are commonly used to compare different categories or groups.

4. Histograms: These plots display the distribution of data by grouping it into intervals or bins and displaying the number of 
   observations in each bin. They are commonly used to explore the distribution of data.

5. Pie Chart: A pie chart is a circular graph that displays the relative size of each category as a slice of the pie. This type of plot is useful for showing proportions and percentages.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

# Note: Also add title, xlabel, and ylabel to the plot.

# ANS:- 

# A scatter plot is a type of plot that displays the relationship between two variables by plotting individual data points. Each data point is represented by a point on the graph, and the position of the point is determined by the values of the two variables.

```python
import matplotlib.pyplot as plt
plt.scatter(x , y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Scatter Plot")
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:


```python
import numpy as np

# For line 1:
x1 = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])

# For line 2: 
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

# For line 3:
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

# For line 4:
x4 = np.array([0, 1, 2, 3, 4, 5])  
y4 = np.array([200, 350, 250, 550, 450, 150])
```

# ANS:- 

# The subplot() function in Matplotlib is used to create multiple plots in the same figure.

# This function allows us to divide the figure into a grid of rows and columns, and then specify which subplot to plot in each cell of the grid.

# The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current subplot.

```python
import matplotlib.pyplot as plt

# Create four line plots in one figure using subplot()
fig , axs = plt.subplots(nrows = 2 , ncols = 2 , figsize = (10,10))

# Plot the first line plot on the top-left subplot
axs[0,0].plot(x1,y1)
axs[0,0].set_title('Line 1')

# Plot the second line plot on the top-right subplot
axs[0,1].plot(x2,y2)
axs[0,1].set_title('Line 2')

# Plot the third line plot on the bottom-left subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot the fourth line plot on the bottom-right subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a common x-axis label and a common y-axis label to the entire figure
fig.text(0.5, 0.04, 'X-axis Label', ha='center')
fig.text(0.04, 0.5, 'Y-axis Label', va='center', rotation='vertical')

plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

<b>ANS :- A bar plot is a type of data visualization that uses rectangular bars to represent data values. In a bar plot, the height or length of each bar corresponds to the value of a particular variable or category. Bar plots can be created horizontally or vertically, and they are commonly used to compare the sizes or values of different categories or variables.</b>

<b>Bar plots are useful for presenting categorical data, such as the frequency of occurrence of different categories or the distribution of a categorical variable.</b>

# Here are some specific reasons why bar plots are used:

<b> 1.Comparing quantities :</b> Bar plots are used to compare the quantities of different categories, such as sales figures, revenue, or counts of items. By visually comparing the heights or lengths of the bars, it's easy to see which category has the largest or smallest value.

<b> 2.Tracking changes over time :</b> Bar plots can also be used to track changes over time within a single category. For example, you might create a bar plot showing the revenue of a single department for each month of the year. By comparing the heights or lengths of the bars for each month, you can see how the revenue changes over time.

<b> 3.Showing relationships :</b> Bar plots can also be used to show relationships between categories. For example, you might create a stacked bar plot showing the sales figures for different products in different regions. By stacking the bars on top of each other, you can see how the sales figures for each product compare across different regions.

<b> 4.Communicating results :</b> Bar plots are a simple and effective way to communicate results to a non-technical audience. By presenting data in a visually appealing format, bar plots can help to convey the main points of your analysis quickly and clearly.

```python
# Plotting Normal Bar plot
import matplotlib.pyplot as plt
plt.bar(company , profit)
plt.xlabel("Company Name")
plt.ylabel("Profit")
plt.title("Normal Bar Plot")
```

```python
# Plotting Horizontal Bar plot
import matplotlib.pyplot as plt
plt.barh(company , profit)
plt.xlabel("Company Name")
plt.ylabel("Profit")
plt.title("Horizontal Bar Plot")
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

<b>ANS:- Box plot is a type of data visualization that displays the distribution of a numerical dataset. The plot is made up of a box and whisker diagram that provides information on the central tendency, spread, and skewness of the data.Box plots are commonly used in statistical analysis, data science, and scientific research to explore and communicate the characteristics of a dataset.</b>

# Box plots are commonly used to visualize the distribution of a dataset and to identify potential outliers or extreme values. They are particularly useful for:
<b>1.Summarizing the distribution of a dataset :</b> Box plots provide a visual summary of the distribution of a dataset based on its five-number summary (minimum, first quartile, median, third quartile, and maximum). This allows us to quickly understand the range, central tendency, and spread of the data.

<b>2.Identifying potential outliers :</b> Box plots can also be used to identify potential outliers or extreme values in a dataset. Any data points that fall outside of the whiskers (i.e., more than 1.5 times the interquartile range from the nearest quartile) are considered outliers and are plotted as individual points. This allows us to quickly identify any unusual or extreme values in the data.

<b>3.Comparing distributions :</b> Box plots can also be used to compare the distribution of two or more datasets. By plotting multiple box plots side by side, we can compare the range, central tendency, and spread of each dataset and identify any differences or similarities between them.

<b>Overall, box plots are a powerful tool for visualizing and summarizing the distribution of a dataset, and are widely used in data analysis, statistical modeling, and data visualization.</b>

```python
import matplotlib.pyplot as plt

box = [box1 , box2]

fig = plt.figure(figsize =(4, 4))

# Creating axes instance
ax = fig.add_axes([0,0,1,1])

# Creating plot
p = ax.boxplot(box)

plt.show()
```
